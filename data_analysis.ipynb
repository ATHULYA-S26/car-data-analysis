## graph analysis

### bar chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
freq=df["selling_price"].value_counts()
freq.plot(kind="bar")
plt.title("frequency of selling price")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
freq=df["mileage"].value_counts()
freq.plot(kind="bar")
plt.title("frequency of mileage")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
freq=df["engine_value"].value_counts().tail(20)
freq.plot(kind="hist")
plt.title("frequency of Engine Type")
```

. most cars are sold at lower price,and very few cars are sold at high prices.
